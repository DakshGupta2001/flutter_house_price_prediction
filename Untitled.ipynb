```python
df2 = df.drop(['area_type','society','availability'],axis='columns')
df2
```

```python
df3['bhk']=df3['size'].apply(lambda x: int(x.split(' ')[0]))
```

```python
df3.head()
```

```python
df3.drop('size',axis=1,inplace=True)
```

```python
df3.head()
```

```python
def convert_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
    
```

```python
df3['sqft']=df3['total_sqft'].apply(convert_to_num)
```

```python
df3
```

```python
df3.drop('total_sqft',axis=1)
```

```python
df3['price_per_sqft'] = df3['price']*100000/df3['sqft']
```

```python
df3
```

```python
df3.describe()
```

```python
location_count=df3.location.value_counts()
```

```python
df3['location']=df3['location'].apply(lambda x: x.strip())
loc_count_less_10=location_count[location_count<=10]
```

```python
loc_count_less_10
```

```python
df3.location = df3.location.apply(lambda x: 'other' if x in loc_count_less_10 else x)
len(df3.location.unique())
```

```python
df3=df3[(df3['sqft']/df3['bhk'])>=300]
```

```python
df3.describe()
```

```python
import numpy as np
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index = True)
    return df_out

df3 = remove_pps_outliers(df3)
df3.shape
```

```python
df3.describe()
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
df3 = remove_bhk_outliers(df3)
# df8 = df7.copy()
df3.shape
```

```python
from sklearn.linear_model import Lasso,LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
from sklearn.compose import make_column_transformer
from sklearn.pipeline import Pipeline
```

```python
col_trans=make_column_transformer((OneHotEncoder(sparse=False,handle_unknown='ignore'),['location']),remainder='passthrough')
```

```python
scaler=StandardScaler()
pipe=Pipeline([('col_trans',make_column_transformer((OneHotEncoder(sparse=False,handle_unknown='ignore'),['location']),remainder='passthrough')),('scaler', StandardScaler()),('decisiontree',DecisionTreeRegressor())])
# pipe2=make_pipeline(col_trans)
pipe
```
